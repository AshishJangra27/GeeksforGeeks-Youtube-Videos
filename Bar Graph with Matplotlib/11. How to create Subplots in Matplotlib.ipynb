### Graph with 2X2

```python
import numpy as np
import matplotlib.pyplot as plt
 
x = np.array([1, 2, 3, 4, 5])

y1 = x**1    # Linear
y2 = x**2    # Square
y3 = x**3    # Cube 
y4 = x**4    # Quad

fig, ax = plt.subplots(2,2, figsize = (8,6))

ax[0,0].plot(x,y1)
ax[0,1].plot(x,y2)
ax[1,0].plot(x,y3)
ax[1,1].plot(x,y4)

ax[0,0].title.set_text("Linear")
ax[0,1].title.set_text("Square")
ax[1,0].title.set_text("Cube")
ax[1,1].title.set_text("Quad")

fig.tight_layout()

plt.show()
```

### Graph with 3X2

```python
import numpy as np
import matplotlib.pyplot as plt
 
x = np.array([1, 2, 3, 4, 5])

y1 = x**1    # Linear
y2 = x**2    # Square
y3 = x**3    # Cube 
y4 = x**4    # Quad
y5 = x**5    # Five
y6 = x**6    # Six

fig, ax = plt.subplots(3,2, figsize = (8,6))

ax[0,0].plot(x,y1)
ax[0,1].plot(x,y2)
ax[1,0].plot(x,y3)
ax[1,1].plot(x,y4)
ax[2,0].plot(x,y5)
ax[2,1].plot(x,y6)

ax[0,0].title.set_text("Linear")
ax[0,1].title.set_text("Square")
ax[1,0].title.set_text("Cube")
ax[1,1].title.set_text("Quad")
ax[2,0].title.set_text("Five")
ax[2,1].title.set_text("Six")

fig.tight_layout()

plt.show()
```
