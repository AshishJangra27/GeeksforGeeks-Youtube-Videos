### Importing Maptlotlib

```python
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
```

### Line Graph

```python
x = ['Ashish','Manish','Manoj']
y = [78, 80,85]

plt.plot(x,y)

plt.show()
```

### Bar Graph

```python
x = ['Ashish','Manish','Manoj']
y = [78, 80,85]

plt.bar(x,y)

plt.show()
```

### Histogram

```python
data = [56,76,56,76,56,57,45,75,66,54]

plt.hist(data)

plt.show()
```

### Scatter Plot

```python
x = [5,2,9,4,7]
y = [9,5,8,4,2]

plt.scatter(x,y)

plt.show()
```
