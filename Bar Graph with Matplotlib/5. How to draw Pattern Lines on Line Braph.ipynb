```python
import matplotlib.pyplot as plt

year  = ['2010', '2011' , '2012' , '2013' , '2014','2015']

apl   = [198 , 199  , 196  , 200  ,  201, 199]
tsl   = [203 , 200  , 190  , 198  ,  204, 200]
amz   = [205 , 195  , 187  , 195  ,  200, 197]
```

```python
plt.plot(year,apl)

plt.show()
```

### Dashed

```python
plt.plot(year,apl,'-.')

plt.show()
```

### Dots

```python
plt.plot(year,apl,'o')

plt.show()
```

```python
plt.plot(year,apl,'o-')

plt.show()
```

```python
plt.plot(year,apl,'g-')

plt.show()
```

```python
plt.plot(year,apl,'r-')

plt.show()
```

```python
plt.plot(year,apl,'r-.')

plt.show()
```

```python
plt.plot(year,apl,'r-.^')

plt.show()
```

```python
plt.plot(year,apl,'b-')
plt.plot(year,tsl,'r-.')
plt.plot(year,amz,'g-')

plt.show()
```

```python
plt.plot(year,apl,'bo')
plt.plot(year,tsl,'ro')
plt.plot(year,amz,'go')

plt.show()
```
