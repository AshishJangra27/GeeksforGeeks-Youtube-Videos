```python
import matplotlib.pyplot as plt

year  = ['2010', '2011' , '2012' , '2013' , '2014']
price = [198 , 199  , 196  , 200  ,  201]
```

```python
plt.plot(year, price)

plt.show()
```

## Adding Labels

```python
plt.plot(year, price)


plt.title('Stock Price in last 5 years')

plt.xlabel('Year')
plt.ylabel('Stock Price')

plt.show()
```
